```python
import pandas as pd

# Read in the three tables
content_df = pd.read_excel("C:\\Users\\lutol\\Downloads\\Content.xlsx")
reaction_df = pd.read_excel("C:\\Users\\lutol\\Downloads\\Reactions.xlsx")
reaction_type_df = pd.read_excel("C:\\Users\\lutol\\Downloads\\ReactionTypes.xlsx")

# Merge the content and reaction tables on the content ID column
merged_df = pd.merge(content_df, reaction_df, on='Content ID')

# Merge the merged table with the reaction type table on the type column
merged_df = pd.merge(merged_df, reaction_type_df, on='Type')

# Save the merged table to a new Excel file
merged_df.to_excel('merged_table.xlsx', index=False)
```

```python
# Read the merged dataset into a pandas dataframe
df = pd.read_csv("C:\\Users\\lutol\\Desktop\\merged_tables.csv")

# Group the dataframe by Category and count the number of occurrences
category_count = df.groupby('Category').size().reset_index(name='Count')

# Sort the dataframe by Count in descending order
category_count_sorted = category_count.sort_values(by='Count', ascending=False)

# Get the top 5 categories
top5_categories = category_count_sorted.head(5)
print(top5_categories
```

```python
import matplotlib.pyplot as plt

# create a bar chart showing the category count
category_counts = df['Category'].value_counts()
plt.bar(category_counts.index, category_counts.values)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Category Distribution')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt

# create a pie chart showing the content type distribution
content_type_count = df['Content '].value_counts()
plt.pie(content_type_count.values, labels=content_type_count.index)
plt.title('Content Type Distribution')

# add a legend with value counts
plt.legend(title="Count", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), labels=['{} - {}'.format(i,j) for i,j in zip(content_type_count.index, content_type_count.values)])
plt.show()

```

```python
import matplotlib.pyplot as plt

# Create a dataframe with the count of each category
category_count = df['Category'].value_counts()

# Get the top 5 categories by count
top_5_categories = category_count[:5]

# Plot the pie chart
plt.pie(top_5_categories.values, labels=top_5_categories.index)
plt.title('Top 5 Categories')

# Add the count values as labels
plt.legend(title="Counts", loc="best", labels=top_5_categories.astype(str), bbox_to_anchor=(1.2, 0.5))

plt.show()

```
